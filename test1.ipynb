```python
output_file=pd.read_csv('Assignment1_Output_table.csv')
```

```python
lot_number=input_file.lot_number.unique()
```

```python
lot_number
```

```python

run=[]
for i in lot_number:
    l=[]
    a=input_file[input_file['lot_number']==i][['panel','Run','group']]
    l.append(i)
    l.append(a['group'].values[0])
    for p in sorted(a.panel.values):
        v=a[['panel','Run']]
        r=v[v['panel']==p]['Run']
        l.append(r.values[0])
    run.append(l)
        
        
        
    
```

```python
output=pd.DataFrame(run)
output=output.drop(8,axis=1)
output
```
